```python
import numpy as np 
from matplotlib import pyplot as plt 
```

```python
def lagrange(x,x_s,y_s):
    return sum([yj*np.prod([(x-xi)/(xj-xi) for xi in x_s if xi!=xj]) for xj,yj in zip(x_s,y_s)])
```

```python
x_s = [70 , 160 , 250 , 350 , 500 , 850]
y_s = [713.556 , 1695.1, 1019.447 , 501.684 , 156.957 , 24.916]
```

```python
plt.plot(x_s, y_s)
plt.show()
```

```python
x = np.linspace(x_s[0], x_s[-1] , num=10 , endpoint=1)
y_inter = [-lagrange(x,x_s , y_s) for x in x]
```

```python
plt.plot(x, y_inter)
plt.plot(x_s,y_s)
plt.show()
```
