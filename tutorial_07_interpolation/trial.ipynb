```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

def gen_points(x):
    return(1/(25*x**2+1))
```

```python
x = np.linspace(-1,1,100,endpoint=True)
y = gen_points(x)
```

```python
x_s = np.append(x[::10],x[len(x)-1])
y_s = np.append(y[::10],y[len(y)-1])
```

```python
x_s = x[::10]
y_s = y[::10]
```

```python

plt.plot(x, y,'-k')
plt.plot(x_s,y_s,'.r')
plt.show()
```

```python
def lag_int(x,x_s,y_s):
    return sum([yj*np.prod([(x-xi)/(xi-xj) for xi in x_s if xi!=xj]) for xj,yj in zip(x_s,y_s)])
```

```python
lx = [lag_int(xp,x_s,y_s) for xp in x]
plt.plot(x, y,'-k')
plt.plot(x_s,y_s,'.r')
plt.plot(x,lx,'-b')
plt.show()
```
